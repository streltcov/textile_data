```python
import numpy
import pandas
import textiles.densities as densities
import seaborn
from prettytable import PrettyTable, ALL
from matplotlib import pyplot
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
########################################################################################################################

table = PrettyTable()
table.hrules = ALL
table.field_names = ['ПАРАМЕТР', 'ЗНАЧЕНИЕ']
table.add_row(["Минимальное значение плотности по основе", densities.warp_min])
table.add_row(["Максимальное значение плотности по основе", densities.warp_max])
table.add_row(["Разброс значение плотности по основе", densities.warp_range])
table.add_row(["Разброс значение плотности по утку", densities.weft_range])
table.add_row(["Среднее значение плотности по основе", densities.warp_mean])
table.add_row(["Среднее значение утку", densities.weft_mean])
table.add_row(["Мода плотности по основе", densities.warp_mode])
table.add_row(["Мода плотности по утку", densities.weft_mode])
table.add_row(["Медиана плотности по основе", densities.warp_median])
table.add_row(["Медиана плотности по утку", densities.weft_median])
print(table)

########################################################################################################################
```

```python
_, ax = pyplot.subplots(nrows=2, ncols=2, figsize=(16, 20))
ax[0][0].hist(densities.warps, bins=15, color='gray')
ax[0][1].hist(densities.wefts, bins=15, color='gray')
ax[1][0].boxplot(densities.warps)
ax[1][1].boxplot(densities.wefts)
pyplot.show()
```

```python
fig = pyplot.figure(figsize=(16, 15))
grid = pyplot.GridSpec(4, 4, hspace=0.5, wspace=0.2)

fig.suptitle('Распределение плотностей нитей по основе и утку', fontsize=18)

ax_main = fig.add_subplot(grid[:-1, :-1])

ax_right = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])
ax_right.invert_yaxis()

ax_bottom = fig.add_subplot(grid[-1, 0:-1], xticklabels=[], yticklabels=[])
ax_bottom.invert_yaxis()

ax_main.scatter(densities.warps, densities.wefts, color='lightcoral')
ax_bottom.hist(densities.warps, 40, color='gray', histtype='stepfilled', orientation='vertical')
ax_right.hist(densities.wefts, 40, color='lightgray', histtype='stepfilled', orientation='horizontal')

pyplot.show()
```
